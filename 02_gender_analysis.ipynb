    # Black Friday Sales Analysis - Gender Analysis

    ## Objective
    Analyze how gender influences customer participation and spending behavior during Black Friday sales.

    ## Steps Performed
    - Examined gender distribution
- Compared purchase patterns by gender
- Visualized gender-wise counts and spending
- Summarized key observations

    ## Insights
    - Gender-based differences can help understand customer behavior
- Spending patterns may vary even when customer counts are different
- This analysis supports targeted marketing decisions


```python
import pandas as pd
```

```python
df = pd.read_csv('BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
for column in df.columns:
    print(df[column].nunique(), "\t: ", column)
```

```python
data = pd.DataFrame({'Ratio' : [len(df[df['Gender'] == 'M']), len(df[df['Gender'] == 'F'])]}, 
                    index = ['Male', 'Female'])

data.plot.pie(y = 'Ratio', figsize = (6,6), autopct = "%.1f")
```

```python
df.groupby('Gender').size().plot(kind = 'pie', 
                                 autopct = "%.1f",
                                 title = 'Gender Ratio',
                                 figsize = (6,6))
```

```python
df.groupby('Gender').size().plot(kind = 'bar', 
                                 figsize = (6,6))
```

```python
df.groupby('Gender').size()
```

```python
df.groupby('Gender')['Purchase'].sum().plot(kind = 'pie', autopct = "%0.1f")
```

```python
df.groupby('Gender')['Purchase'].mean().plot(kind = 'pie', autopct = "%0.1f")
```

## Extra Visuals
Additional visuals for clearer gender-wise comparison.

```python
import matplotlib.pyplot as plt
import seaborn as sns

if 'Gender' in df.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(x='Gender', data=df)
    plt.title("Customer Count by Gender")
    plt.show()

    if 'Purchase' in df.columns:
        plt.figure(figsize=(6,4))
        df.groupby('Gender')['Purchase'].mean().plot(kind='bar')
        plt.title("Average Purchase by Gender")
        plt.ylabel("Average Purchase")
        plt.show()
```
