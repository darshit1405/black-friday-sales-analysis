## Objective

## Business Questions
- Which customer groups spend more during Black Friday?
- Which demographics contribute the most to revenue?
- Which product-related features seem most important?
- How can these insights support business decisions?


```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/BlackFriday.csv')
print("Shape:", df.shape)
print(df.head())
```

## Key Insights
- Customer spending behavior is influenced by demographic as well as product-related factors.
- Gender, age, marital status, and occupation all add useful business context.
- Product categories also affect purchasing patterns.
- Multi-column analysis gives deeper insight than one-variable analysis.


```
plt.figure(figsize=(8,4))
sns.histplot(df['Purchase'], bins=30, kde=True)
plt.title('Purchase Distribution')
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x='Gender', y='Purchase', data=df, estimator='mean')
plt.title('Average Purchase by Gender')
plt.show()
```

## Final Conclusion
Black Friday sales are shaped by multiple customer and product characteristics.  
A structured EDA helps identify stronger customer segments, better marketing directions, and useful inputs for future predictive modeling.
