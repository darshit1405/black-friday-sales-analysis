## Extra Visuals
Additional visuals for occupation and product analysis.

```python
import matplotlib.pyplot as plt
import seaborn as sns

if 'Occupation' in df.columns:
    plt.figure(figsize=(10,4))
    sns.countplot(x='Occupation', data=df)
    plt.title("Customer Count by Occupation")
    plt.show()

    if 'Purchase' in df.columns:
        plt.figure(figsize=(10,4))
        df.groupby('Occupation')['Purchase'].mean().plot(kind='bar')
        plt.title("Average Purchase by Occupation")
        plt.ylabel("Average Purchase")
        plt.show()

for col in ['Product_Category_1', 'Product_Category_2', 'Product_Category_3']:
    if col in df.columns:
        plt.figure(figsize=(10,4))
        sns.countplot(x=col, data=df, order=df[col].value_counts().index[:10])
        plt.title(f"Top Values in {col}")
        plt.show()
```
